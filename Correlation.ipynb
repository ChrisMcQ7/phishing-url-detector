```python
from IPython.display import display, HTML
display(HTML("<style>.output_scroll { overflow-x: visible !important; }</style>"))

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("phishing_features.csv")

# Optional: drop non-numeric or irrelevant columns
# df = df.drop(columns=["url", "domain", ...])  # Keep only numerical features

# Calculate correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Features")
plt.show()
```
